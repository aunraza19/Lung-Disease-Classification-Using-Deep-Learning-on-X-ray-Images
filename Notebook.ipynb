```python
import numpy as np
import pandas as pd
from pathlib import Path
import os.path
import matplotlib.pyplot as plt
from IPython.display import Image, display
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.models import Model
#from tensorflow.keras.optimizers import Adam

# --- Multi-GPU Implementation ---
strategy = tf.distribute.MirroredStrategy()
print('Number of devices: {}'.format(strategy.num_replicas_in_sync))

image_dir = Path('/kaggle/input/dataset/data_lung')

filepaths = list(image_dir.glob(r'**/*.jpg')) + list(image_dir.glob(r"**/**/*.png"))
labels = list(map(lambda x: os.path.split(os.path.split(x)[0])[1], filepaths))

filepaths = pd.Series(filepaths, name='Filepath').astype(str)
labels = pd.Series(labels, name='Label')

image_df = pd.concat([filepaths, labels], axis=1)

image_df = image_df[image_df['Label'].apply(lambda x: x[-2:] != 'GT')]

image_df = image_df.sample(frac=1).reset_index(drop=True)

image_df['dataset'] = image_df['Filepath'].apply(lambda x: x.split('/')[-3])

train_df = image_df[image_df['dataset']=='train']
test_df = image_df[image_df['dataset']=='test']

# Correct VGG16 preprocessing function
train_generator = tf.keras.preprocessing.image.ImageDataGenerator(
 preprocessing_function=tf.keras.applications.vgg16.preprocess_input,
 validation_split=0.2
)
test_generator = tf.keras.preprocessing.image.ImageDataGenerator(
 preprocessing_function=tf.keras.applications.vgg16.preprocess_input
)

train_images = train_generator.flow_from_dataframe(
 dataframe=train_df,
 x_col='Filepath',
 y_col='Label',
 target_size=(224, 224), 
 color_mode='rgb',
 class_mode='categorical',
 batch_size=32 * strategy.num_replicas_in_sync, # Scale batch size
 shuffle=True,
 seed=42,
 subset='training'
)
val_images = train_generator.flow_from_dataframe(
 dataframe=train_df,
 x_col='Filepath',
 y_col='Label',
 target_size=(224, 224),  
 color_mode='rgb',
 class_mode='categorical',
 batch_size=32 * strategy.num_replicas_in_sync, # Scale batch size
 shuffle=True,
 seed=42,
 subset='validation'
)
test_images = test_generator.flow_from_dataframe(
 dataframe=test_df,
 x_col='Filepath',
 y_col='Label',
 target_size=(224, 224),  
 color_mode='rgb',
 class_mode='categorical',
 batch_size=32,
 shuffle=False
)

# --- Model Creation within the strategy scope ---
with strategy.scope():
    pretrained_model = VGG16(
    include_top=False,
    weights='imagenet',
    input_shape=(224, 224, 3)
    )
    pretrained_model.trainable = False
    pretrained_model.summary()
    
    output = pretrained_model.output
    
    # Apply the new Conv2d block to the output
    x = tf.keras.layers.Conv2D(filters=512, kernel_size=3, padding='same', activation='relu', name='block6_conv1')(output)
    x = tf.keras.layers.Conv2D(filters=512, kernel_size=3, padding='same', activation='relu', name='block6_conv2')(x)
    x = tf.keras.layers.Conv2D(filters=512, kernel_size=3, padding='same', activation='relu', name='block6_conv3')(x)
    x = tf.keras.layers.GlobalAveragePooling2D(name='block6_gap')(x)
    x = tf.keras.layers.Flatten(name='block7_flatten')(x)
    x = tf.keras.layers.Dense(10, activation='softmax')(x)
    
    # Create a new model with the combined layers
    model = tf.keras.Model(inputs=pretrained_model.input, outputs=x, name='LungNet22')
    model.summary()
    #adam_optimizer = Adam(learning_rate=0.000001)
    
    model.compile(
    #    optimizer=adam_optimizer, # Pass the optimizer object here
        optimizer='Adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    
history = model.fit(
 train_images,
 validation_data=val_images,
 epochs=15,
 callbacks=[
     tf.keras.callbacks.EarlyStopping(
         monitor='val_loss',
         patience=10,
         restore_best_weights=True
     )
 ]
)

model.save('VGG16&LungNet22.keras')

pd.DataFrame(history.history)[['accuracy', 'val_accuracy']].plot()
plt.title("Accuracy")
plt.show()


pd.DataFrame(history.history)[['loss', 'val_loss']].plot()
plt.title("Loss")
plt.show()


results = model.evaluate(test_images, verbose=0)
print("    Test Loss: {:.5f}".format(results[0]))
print("Test Accuracy: {:.2f}%".format(results[1] * 100))

pred = model.predict(test_images)
pred = np.argmax(pred, axis=1)
labels = (train_images.class_indices)
labels = dict((v, k) for k, v in labels.items())
pred = [labels[k] for k in pred]
print(f'The first 6 predictions: {pred[:6]}')

from sklearn.metrics import classification_report

y_test = list(test_df.Label)
print(classification_report(y_test, pred))

fig, axes = plt.subplots(nrows=3, ncols=5, figsize=(15, 7),
                      subplot_kw={'xticks': [], 'yticks': []})
for i, ax in enumerate(axes.flat):
 ax.imshow(plt.imread(test_df.Filepath.iloc[i]))
 ax.set_title(f"True: {test_df.Label.iloc[i]}\nPredicted: {pred[i]}")
plt.tight_layout()
plt.show()
```

```python
# inference_pipeline_lungnet22.py
# Run with: python inference_pipeline_lungnet22.py
# Requirements: opencv-python, numpy, matplotlib, tensorflow (and torch only if you want to load EnsNet PyTorch model)

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# Optional: torch used only if you want to load a PyTorch EnsNet model
try:
    import torch
    TORCH_AVAILABLE = True
except Exception:
    TORCH_AVAILABLE = False

# ---------------------------
# Helper: load Keras classifier
# ---------------------------
def load_classifier(model_path):
    if not os.path.exists(model_path):
        raise FileNotFoundError(f"classifier not found: {model_path}")
    model = load_model(model_path)
    print("Loaded classifier:", model_path)
    return model

# ---------------------------
# 1) Autoencoder denoising
# ---------------------------
def denoise_with_autoencoder(img_gray, autoenc_path=None):
    """
    img_gray: uint8 grayscale image (H,W)
    autoenc_path: path to a keras autoencoder .h5 (optional)
    Returns uint8 denoised grayscale (H,W)
    """
    if autoenc_path and os.path.exists(autoenc_path):
        try:
            ae = load_model(autoenc_path)
            # prepare input for AE: model may expect (H,W,1) or (H,W,3). We assume 1-channel AE.
            x = img_gray.astype("float32") / 255.0
            x = np.expand_dims(x, axis=-1)   # (H,W,1)
            x = np.expand_dims(x, axis=0)    # (1,H,W,1)
            den = ae.predict(x)[0,...]
            den = (np.clip(den, 0, 1) * 255).astype(np.uint8)
            if den.shape[-1] == 1:
                den = den[...,0]
            return den
        except Exception as e:
            print("Autoencoder load/predict failed; falling back to OpenCV denoise. Error:", e)

    # Fallback: OpenCV fastNlMeansDenoising for grayscale
    denoised = cv2.fastNlMeansDenoising(img_gray, None, h=10, templateWindowSize=7, searchWindowSize=21)
    return denoised

# ---------------------------
# 2) Annotation removal (EnsNet or fallback)
# ---------------------------
def remove_annotations(img_gray, ensnet_path=None):
    """
    img_gray: uint8 grayscale
    ensnet_path: optional path to EnsNet or inpainting model (if None, uses OpenCV inpainting)
    Returns uint8 grayscale
    """
    # If user has a PyTorch EnsNet model (common implementation), try to use it.
    if ensnet_path and os.path.exists(ensnet_path) and TORCH_AVAILABLE:
        try:
            # NOTE: real EnsNet inference typically requires the architecture code.
            # Here we attempt a minimal torch.load inference if the model is a full scripted module.
            model = torch.jit.load(ensnet_path, map_location='cpu') if ensnet_path.endswith('.pt') or ensnet_path.endswith('.jit') else torch.load(ensnet_path, map_location='cpu')
            model.eval()
            x = img_gray.astype('float32') / 255.0
            x = np.expand_dims(x, axis=0)   # 1,H,W
            x = np.expand_dims(x, axis=0)   # 1,1,H,W  (N,C,H,W)
            with torch.no_grad():
                t = torch.from_numpy(x)
                out = model(t)
                out_np = out.squeeze().cpu().numpy()
                out_np = (np.clip(out_np,0,1)*255).astype(np.uint8)
                if out_np.ndim == 3 and out_np.shape[0] == 1:
                    out_np = out_np[0]
                return out_np
        except Exception as e:
            print("Could not run EnsNet model (or model not compatible). Falling back to morphological inpainting. Error:", e)

    # Fallback: detect bright annotations (text/markers) and inpaint
    # 1) Binarize to find bright text (tweak threshold if needed)
    _, mask = cv2.threshold(img_gray, 200, 255, cv2.THRESH_BINARY)
    # 2) Morphology to expand mask
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7,7))
    mask = cv2.dilate(mask, kernel, iterations=2)
    # 3) Inpaint (Telea works well)
    inpainted = cv2.inpaint(img_gray, mask, inpaintRadius=3, flags=cv2.INPAINT_TELEA)
    return inpainted

# ---------------------------
# 3) CLAHE (RGB-safe)
# ---------------------------
def apply_clahe_rgb(img_rgb, clipLimit=3.0, tileGridSize=(10,10)):
    """
    img_rgb: uint8 RGB image
    returns uint8 RGB image after CLAHE applied on L channel (LAB space).
    """
    lab = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2LAB)
    l, a, b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize=tileGridSize)
    l_cl = clahe.apply(l)
    lab_cl = cv2.merge((l_cl, a, b))
    rgb_cl = cv2.cvtColor(lab_cl, cv2.COLOR_LAB2RGB)
    return rgb_cl

# ---------------------------
# 4) Green Fire Blue LUT / pseudocolor
# ---------------------------
def apply_green_fire_blue_lut(img_gray, lut_path=None):
    """
    img_gray: uint8 grayscale (H,W)
    lut_path: optional path to an ImageJ .lut file (Green Fire Blue.lut). If provided, we attempt to parse it.
    Returns uint8 RGB (H,W,3) with LUT applied.
    """
    # Try to parse .lut (ImageJ format)
    if lut_path and os.path.exists(lut_path):
        try:
            with open(lut_path, 'rb') as f:
                data = f.read()
            # ImageJ LUTs can be binary or ascii; a robust parser is non-trivial.
            # Try a simple approach: if the file contains ASCII lines with numbers, parse them.
            text = None
            try:
                text = data.decode('utf-8')
            except Exception:
                text = None
            if text:
                vals = []
                for line in text.splitlines():
                    parts = [p.strip() for p in line.split() if p.strip()]
                    if len(parts) >= 3:
                        try:
                            r,g,b = int(parts[0]), int(parts[1]), int(parts[2])
                            vals.append((r,g,b))
                        except:
                            continue
            else:
                vals = []
            if len(vals) >= 256:
                lut = np.array(vals[:256], dtype=np.uint8)  # (256,3)
                normalized = img_gray.clip(0,255).astype(np.uint8)
                rgb = lut[normalized]
                return rgb
        except Exception as e:
            print("Failed to parse LUT file; falling back to synthetic colormap. Error:", e)

    # Fallback: construct a "Green-Fire-Blue-like" colormap using matplotlib
    # (Green -> Yellow/Orange -> Red -> Purple -> Blue)
    import matplotlib
    from matplotlib.colors import LinearSegmentedColormap
    colors = ['#00A000', '#FFFF66', '#FF4500', '#800080', '#0000FF']
    cmap = LinearSegmentedColormap.from_list('green_fire_blue', colors, N=256)
    normalized = img_gray.astype(np.float32) / 255.0
    mapped = cmap(normalized)[:, :, :3]  # (H,W,3) floats 0..1
    rgb = (mapped * 255).astype(np.uint8)
    return rgb

# ---------------------------
# Pipeline: full preprocessing for LungNet22
# ---------------------------
def preprocess_for_lungnet22(image_path,
                             autoenc_path=None,
                             ensnet_path=None,
                             greenfire_lut_path=None,
                             target_size=(224,224)):
    """
    Runs:
      1) read grayscale
      2) autoencoder denoise
      3) annotation removal
      4) CLAHE enhancement
      5) Green-Fire-Blue LUT -> RGB
      6) resize -> 224x224
      7) normalize to [0,1]
    Returns preprocessed array shaped (1, H, W, 3) dtype float32
    """
    if not os.path.exists(image_path):
        raise FileNotFoundError(f"image not found: {image_path}")

    # 1) read grayscale (most X-rays are single-channel)
    img_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img_gray is None:
        raise ValueError("Failed to read image (is the path correct and is it an image?).")

    # 2) Denoise with AE or fallback
    den = denoise_with_autoencoder(img_gray, autoenc_path=autoenc_path)

    # 3) Remove annotations
    clean = remove_annotations(den, ensnet_path=ensnet_path)

    # 4) We currently have a clean grayscale image; convert to RGB before CLAHE (CLAHE on L channel)
    rgb_before_clahe = cv2.cvtColor(clean, cv2.COLOR_GRAY2RGB)

    # 5) CLAHE enhancement (on LAB L channel) — uses your clipLimit and tileGridSize
    clahe_rgb = apply_clahe_rgb(rgb_before_clahe, clipLimit=3.0, tileGridSize=(10,10))

    # 6) Apply Green Fire Blue LUT (works from grayscale; we pass the CLAHE L-channel as grayscale input)
    # Extract L channel to pass to LUT
    lab_tmp = cv2.cvtColor(clahe_rgb, cv2.COLOR_RGB2LAB)
    l_channel = lab_tmp[:,:,0]
    gfb_rgb = apply_green_fire_blue_lut(l_channel, lut_path=greenfire_lut_path)

    # 7) Resize to model input size (224x224)
    resized = cv2.resize(gfb_rgb, target_size, interpolation=cv2.INTER_AREA)

    # 8) Normalize to [0,1] float32 (common for Keras models)
    x = resized.astype('float32') / 255.0
    x = np.expand_dims(x, axis=0)  # (1, H, W, 3)
    return x, {
        "denoised": den,
        "annotation_removed": clean,
        "clahe_rgb": clahe_rgb,
        "gfb_rgb_full": gfb_rgb,
        "resized_rgb": resized
    }

# ---------------------------
# Prediction wrapper
# ---------------------------
def predict_lungnet(model, preprocessed_x, class_names=None):
    """
    model: Keras model
    preprocessed_x: (1,224,224,3)
    class_names: list of 10 class names in same order as model output (optional)
    """
    pred = model.predict(preprocessed_x)
    # assume softmax over 10 classes
    idx = int(np.argmax(pred, axis=1)[0])
    prob = float(pred[0, idx])
    label = class_names[idx] if class_names is not None and len(class_names) > idx else f"class_{idx}"
    return {"label": label, "index": idx, "probability": prob, "raw_preds": pred}

# ---------------------------
# Example usage
# ---------------------------
if __name__ == "__main__":
    # --- UPDATE THESE PATHS ---
    lungnet_model_path = "/kaggle/working/VGG16&LungNet22.keras"        # your trained LungNet22 Keras model
    autoenc_path = "autoencoder_denoiser.h5"        # optional: your autoencoder model
    ensnet_path = "ensnet_inference.pth"             # optional: your EnsNet model (PyTorch scripted/.pt recommended)
    greenfire_lut_path = "Green Fire Blue.lut"      # optional: ImageJ LUT file
    test_image = "/kaggle/input/example-img/mass00002.jpg"       # external single X-ray you want to infer
    # --------------------------

    # load classifier (LungNet22)
    model = load_classifier(lungnet_model_path)

    # Preprocess (applies the 4-step pipeline)
    x, intermediates = preprocess_for_lungnet22(
        test_image,
        autoenc_path=autoenc_path if os.path.exists(autoenc_path) else None,
        ensnet_path=ensnet_path if os.path.exists(ensnet_path) else None,
        greenfire_lut_path=greenfire_lut_path if os.path.exists(greenfire_lut_path) else None,
        target_size=(224,224)
    )

    # Define class names (replace with your actual mapping)
    class_names =  [
            'control', 'covid', 'effusion', 'lung Opacity', 'mass',
            'nodule', 'pneumonia', 'pneumothorax', 'pulmunary fibrosis', 'tuberculosis'
        ]
    # Predict
    result = predict_lungnet(model, x, class_names=class_names)
    print("Prediction:", result)

    # Optional: show intermediate steps
    plt.figure(figsize=(12,6))
    plt.subplot(1,4,1); plt.imshow(intermediates['denoised'], cmap='gray'); plt.title('Denoised'); plt.axis('off')
    plt.subplot(1,4,2); plt.imshow(intermediates['annotation_removed'], cmap='gray'); plt.title('Annotations removed'); plt.axis('off')
    plt.subplot(1,4,3); plt.imshow(intermediates['clahe_rgb']); plt.title('CLAHE (RGB)'); plt.axis('off')
    plt.subplot(1,4,4); plt.imshow(intermediates['resized_rgb']); plt.title('Final resized (RGB)'); plt.axis('off')
    plt.show()
```
